```python
import requests
lat = float(input("Enter the latitude of your region: "))
lon = float(input("Enter the longitude of your region: "))
url = "http://api.openweathermap.org/data/2.5/air_pollution"
api = "YOUR API KEY"
parameters = {
    'lat' : lat,
    'lon' : lon,
    'appid' : api
}
response = requests.get(url = url,params = parameters)
response_json = response.json()
```

```python
components = response_json['list'][0]['components']
co = (components['co'])/1000
aqi = []
```

```python
def aqi_calculator(conc,bphi,bplow,aqihi,aqilow):
    aqi = ((aqihi - aqilow)/(bphi - bplow)) * (conc - bplow) + aqilow
    return aqi
```

```python
aqi_no2 = 0
if 0 <= components['no2'] <= 40:
   aqi_no2 = aqi_calculator(components['no2'],40,0,50,0)
elif 41 <= components['no2'] <= 80:
   aqi_no2 = aqi_calculator(components['no2'],80,41,100,51)
elif 81 <= components['no2'] <=  180:
   aqi_no2 = aqi_calculator(components['no2'],180,81,200,101)
elif 181 <= components['no2'] <= 280:
   aqi_no2 = aqi_calculator(components['no2'],280,181,300,201)
elif 281 <= components['no2'] <= 400:
   aqi_no2 = aqi_calculator(components['no2'],400,281,400,301)
aqi.append(aqi_no2)
```
